```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_thread = np.array([1,2,3,4,5,6,7,8])
my_time = np.array([52, 49, 52, 52, 53, 58, 60, 60])

def new_func(my_thread, my_time):
    plt.bar(my_thread, my_time, color='blue')
    plt.xlabel('Number of threads')
    plt.ylabel('Time (s)')
    plt.title('Time vs Number of threads')
    plt.ylim(48, 65)
    plt.grid()
    plt.show()

new_func(my_thread, my_time)
```
